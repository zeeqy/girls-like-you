```python
import numpy as np
import os
from scipy import stats
import matplotlib.pyplot as plt
```

```python
cur_dir = os.path.join(os.getcwd())
data_path = os.path.join(cur_dir, "data", "ks.csv")
fin = open(data_path)
lst = []
for line in fin:
    lst.append(float(line.strip('\n')))
```

```python
stats.kstest(lst, 'uniform')
```

```python
sorted_lst = np.sort(lst)
p = 1. * np.arange(len(lst)) / (len(lst) - 1)
plt.figure(figsize=(8,6))
plt.plot(sorted_lst, p, color="blue", linestyle = '-', linewidth=2.0)
plt.plot([0,1], [0,1], color="red", linestyle = ':', linewidth=4.0)
plt.xticks([0,1],['0','$|\ J\ |$'])
plt.legend(['Sampled', 'Expected'],loc='upper left',fontsize=15.5)
plt.ylabel('p',fontsize=20)
plt.xlabel('join results',fontsize=20)
# plt.show()
plt.savefig('ks_eo.eps', format='eps', dpi=1200)
```
